```python
import math
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
normal = norm(0, math.sqrt(9)) 
print(normal.pdf(4))
print(normal.cdf(2))       
print(normal.rvs()) 
fig, ax = plt.subplots(1, 1)
x = np.linspace(-3,3,1000)
y = norm.pdf(x)
ax.plot(x,y)
plt.show()
```
